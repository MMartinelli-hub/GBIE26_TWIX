```python
import json

retrieved_definitions = json.load(open("uri_retrieved_definitions.json", "r", encoding="utf-8"))

with open("uri_retrieved_definitions_el_format.jsonl", "w", encoding="utf-8") as f:
    for id, content in retrieved_definitions.items():
        names = content["names"]
        definitions = content["definitions"]
        first_name = names[0] if len(names) > 0 else ""
        synonyms = ""
        if len(names) > 1:
            synonyms = ", ".join(names[1:])
            synonyms = f"Sinonimi: {synonyms}"
        if not first_name:
            print(f"Warning: No name found for ID {id}")
        first_definition = definitions[0] if len(definitions) > 0 else ""
        if synonyms:
            definition = f"{first_definition} {synonyms}"
        else:
            definition = first_definition
        entry = {
            "id": id,
            "name": first_name,
            "description": definition
        }
        f.write(json.dumps(entry, ensure_ascii=False) + "\n")
```
